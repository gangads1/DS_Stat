```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Continuous variable

```python
import scipy.stats as stats

mu = 20  ## mean value
sigma = 2  ## SD
h = sorted(np.random.normal(mu, sigma, 100))
```

```python
len(h)
```

```python
np.mean(h)
```

```python
## for verification, if our mean value and SD value is varying according to the normal distribution as well
abs(20 - np.mean(h)) < 2
```

```python
# bins = 30
# h is the dataset


plt.hist(h, 30, density=True)
```

As we can see the data is continuous.

```python
plt.figure(figsize=(10,5))

fit = stats.norm.pdf(h, np.mean(h), np.std(h))

plt.plot(h, fit, '-o')

plt.hist(h, density=True)
```

### Categorical Data

```python
# group type of juice which will be using and these are the ounces which each of them will produce

df = pd.DataFrame({'group':['a','a','a','b','b','b','c','c','c'], 'ounces':[4,3,12,6,7.5,8,3,5,6]})

df
```

```python
plt.hist(df.group)
```

```python
# Nominal Data

df['rating'] = np.random.randint(0,5,9)
df
```

```python
data_pie = df.groupby(['group'], axis=0)['ounces'].sum()

plt.pie(np.array(data_pie), labels=('a', 'b', 'c'), explode=(0, 0.1, 0), shadow=True, autopct='%d%%')
```
